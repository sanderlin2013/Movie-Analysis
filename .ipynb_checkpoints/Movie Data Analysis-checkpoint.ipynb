## Initial Viewing of Datasets

### IMDB Dataset

First we are going to `import sqlite3`. Then we will connect to the dataset and find the datasets table names. 

```python
#Importing necessary modules:
import sqlite3
import pandas as pd
```

```python
#Create connection to the database and cursor object:
conn = sqlite3.connect('data/im.db')
cur = conn.cursor()
#Find table names
cur.execute("""SELECT name FROM sqlite_master WHERE type = 'table';""")
# Fetch the result and store it in table_names
table_names = cur.fetchall()
table_names
```

Above we see 8 tables names - lets look into those tables now.

For the movie_basics table:

```python
# Getting all the values in the movie_basics table
cur.execute("""SELECT * FROM movie_basics;""");
```

```python
#creating movie_basics dataframe
imbd_movie_basics = pd.DataFrame(
    data=cur.execute("""SELECT * FROM movie_basics;""").fetchall(),
    columns=[x[0] for x in cur.description]
)
imbd_movie_basics
```

For the directors table:

```python
#Getting data and creating the dataframe for directors table
cur.execute("""SELECT * FROM directors;""")
imbd_directors = pd.DataFrame(
    data=cur.execute("""SELECT * FROM directors;""").fetchall(),
    columns=[x[0] for x in cur.description]
)
imbd_directors
```
